```python
#Caesar шифрование одного числа
def encrypt(m, key):
    c = (m + key) % 256
    return c
```

```python
#Caesar дешифрование одного числа
def decrypt(m, key):
    c = (m - key) % 256
    return c
```

```python
#Шифрование массива чисел
def encrypt_data(data, key):
    cypher_data = []
    for m in data:
        c = encrypt(m, key)
        cypher_data.append(c)
    return cypher_data
```

```python
#Дешифрование массива чисел
def decrypt_data(data_c, key):
    data = []
    for c in data_c:
        m = decrypt(c, key)
        data.append(m)
    return data
```

Задание 5.
Расшифровать файл ff2_affine_c_all.bmp. Шифр аффинный. a=167, b=
35. Зашифровать, оставив первые 50 байт без изменения. Сравнить с
оригинальным изображением.

```python
def gcd(a, b):
    while a != 0:
        a, b = b % a, a
    return b


def findModInverse(a, m):
    # Returns the modular inverse of a % m, which is
    # the number x such that a*x % m = 1
    if gcd(a, m) != 1:
        return None # no mod inverse if a & m aren't relatively prime
    # Calculate using the Extended Euclidean Algorithm:
    u1, u2, u3 = 1, 0, a
    v1, v2, v3 = 0, 1, m
    while v3 != 0:
        q = u3 // v3 # // is the integer division operator
        v1, v2, v3, u1, u2, u3 = (u1 - q * v1), (u2 - q * v2), (u3 - q * v3), v1, v2, v3
    return u1 % m


def ex5():
    a = 19
    b = 236
    l = 256

    a_1 = findModInverse(a, l)
    if (a_1 == None):
        print('Error! findModInverse error!', a)
        return

    encrypt_data_arr = rwf.read_data_1byte('ff2_affine_c_all.bmp')
    decrypt_data_arr = []
    for e in encrypt_data_arr:
        m = (e - b) * a_1 % l
        decrypt_data_arr.append(m)
    rwf.write_data_1byte('ff2_affine_c_all_decrypt.bmp', decrypt_data_arr)

    data_arr = rwf.read_data_1byte('ff2_affine_c_all_decrypt.bmp')
    crypt_arr = []
    for i in range(50):
        crypt_arr.append(data_arr[i])
    for i in range(50, len(data_arr)):
        e = (data_arr[i] + b) * a % l
        crypt_arr.append(e)
    rwf.write_data_1byte('ff2_affine_c_all_crypt50.bmp', crypt_arr)


```

Задание 6. 
Дешифровать файл text10_affine_c_all.txt. Шифр аффинный.
Подсчитать сколько перебрали вариантов ключей.

```python
def ex6():
    a = 1
    b = 1
    l = 256
    count = 0
    for a in range(1000):
        a_1 = findModInverse(a, l)
        if (a_1 == None):
            count += 1
            continue
        for b in range(1000):
            encrypt_data_arr = rwf.read_data_1byte('text10_affine_c_all.txt')
            decrypt_data_arr = []
            for e in encrypt_data_arr:
                m = ((e - b) * a_1) % l
                decrypt_data_arr.append(m)
            txt = ''.join([chr(s) for s in decrypt_data_arr])
            # print('decrypt_data =', txt)
            is_english = det_eng.isEnglish(txt)
            count += 1
            if (is_english):
                print("Decrypted with a =", a, "and b =", b)
                print("Count =", count)
                print("True count =", a * b)
                rwf.write_data_1byte('text10_affine_c_all_decrypt.txt', decrypt_data_arr)
                return count
```

Задание 8. Шифр Вижинера. Ключ magistr

```python
def ex8():
    encrypt_data_arr = rwf.read_data_1byte('im6_vigener_c_all.bmp')
    key = 'magistr'
    k = [ord(s) for s in key]
    decrypt_data_arr = []
    j = 0 #перебор ключа
    l = 256 #размер алфавита
    for i in range(len(encrypt_data_arr)):
        j = j % len(k)
        m = (encrypt_data_arr[i] - k[j]) % l
        decrypt_data_arr.append(m)
        j += 1
    rwf.write_data_1byte('im6_vigener_c_all_decrypt.bmp', decrypt_data_arr)

    data_arr = rwf.read_data_1byte('im6_vigener_c_all_decrypt.bmp')
    crypt_arr = []
    j = 0
    for i in range(50):
        crypt_arr.append(data_arr[i])
    for i in range(50, len(data_arr)):
        j = j % len(k)
        e = (data_arr[i] + k[j]) % l
        crypt_arr.append(e)
        j += 1
    rwf.write_data_1byte('im6_vigener_c_all_crypt50.bmp', crypt_arr)
```
